```python
import sys
sys.path.append('..')
from Final_Models import Final_Net
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
network = Final_Net(2, 36, 100, 20, 20, 32).to('cuda')
network.load_state_dict(torch.load('../../Model3_Training/Attempt1/Model_2.pt'))
network.eval()

features = torch.tensor(torch.load('Data/MSA_tensors/T1038-D1_MSA.pt'))
dist_mat = torch.load('Data/distance_matrices/T1038-D1_distmat.pt')
features = torch.nn.functional.one_hot(features[None], num_classes=22).float().permute(0,2, 1,3)

out = torch.bucketize(dist_mat, torch.tensor([2,8]))
out = out[:,:]
out[out == 0] = -100
out -= 1

features = features[:,:, :200]
features.shape

out[out== -101]= 0
out = 1 - out
```

```python
results = pd.DataFrame()



mask = torch.ones_like(out)

for j in range(0):
    torch.diagonal(mask,offset=j).fill_(0)
    torch.diagonal(mask,offset=-j).fill_(0)
    
print((out*mask).sum())

with torch.no_grad():
    preds = network(features.to('cuda'), 20, mask=None, intermediates=True)
    
for i in range(1,20):

    pred = preds[i][0]
    pred = (pred.permute(0,2,1) + pred)
    pred = torch.softmax(pred, dim=0)
    
    pred = pred[0].to('cpu')    
    #pred = (pred + pred.T)/2

    pred = pred*mask
    #min_ = torch.topk(pred.flatten(), 114//5)[0][-1]

    min_ = 0.5
    pred_cont = (pred >= min_).long()
    
    x = ((((pred_cont) * (out))*mask).sum() / ((out)*mask).sum()).item()
    y =  ((((pred_cont) * (out))*mask).sum() / ((pred_cont)*mask).sum()).item()
    if x == 0 and y == 0:
        f1 = None
    else:
        f1 = 2*(x*y)/(x+y)

    
    results = results.append({'Recall': x ,'Presicion': y, 'f1': f1}, ignore_index=True)

```
